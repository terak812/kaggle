### PreProcessing

```python
def drop_col(train, percent):
    drop_col = []
    for name_col, value in train.isnull().sum().items():
        value = value/train.shape[0]*100
        if value >= percent:
            drop_col.append(name_col)
    return drop_col
```

```python
drop_col = drop_col(train, 80)
```

```python
train.columns
```

```python
def pre_process(train, drop_col):
    for i in train.columns:
        if i == 'Name':
            train = train.drop(i, axis=1)
    
    train = train.drop(drop_col, axis=1)
    train = train.interpolate()

    # Find categorical columns to encode 
    objects = [f for f in train.columns if train.dtypes[f] == 'object']

    from sklearn.preprocessing import LabelEncoder
    le = LabelEncoder()
    # Encode labels for object columns
    for value in train[objects]:
        train[value] = le.fit_transform(train[value])
    return train 
```

```python
y = train[train.columns[-1]]
# Drop 'Depression' column
X = train.drop(train.columns[-1], axis=1)
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
from sklearn.pipeline import Pipeline
inputs = [('Pre-Process', pre_process(train, drop_col)),
          ('model', GaussianNB())]

pipe = Pipeline(inputs)
pipe
```

```python
model.fit(X_train, y_train)
y_pred = model.predict(X_valid)
```

```python
# check train, valid score
train_score = model.score(X_train, y_train)
valid_score = model.score(X_valid, y_valid)
print(f'train_score= {train_score*100} %')
print(f'valid_score= {valid_score*100} %')
```
